```python
reviews = []

def getReview(id):
    r = requests.get(f'https://www.imdb.com/title/{id}/criticreviews/', headers=HEADERS)
    soup = BeautifulSoup(r.text, 'html.parser')
    json_data = soup.find('script', {'id': '__NEXT_DATA__'}).string
    data = json.loads(json_data)
    x = data['props']['pageProps']['contentData']['section']
    # append with id as key and reviews as value as a json object
    reviews.append({id: x})

# Convert reviews to JSON format
# reviews_json = json.dumps(reviews, indent=4)

```

```python
REVIEWS=[]
for i in range(250):
    for attempt in range(2):  # Try twice
        try:
            r = requests.get(f'https://www.imdb.com/title/{df['imdbID'][i]}/criticreviews/', headers=HEADERS, timeout=5)
            soup = BeautifulSoup(r.text, 'html.parser')
            json_data = soup.find('script', {'id': '__NEXT_DATA__'}).string
            data = json.loads(json_data)
            x = data['props']['pageProps']['contentData']['section']
            # append with id as key and reviews as value as a json object
            REVIEWS.append({df['Title'][i]: x})
            print(f'{i}: {df['Title'][i]} done')
            break  # If the request is successful, break the loop
        except requests.exceptions.RequestException as e:
            if attempt < 1:  # If this was the first attempt, continue to the next attempt
                continue
            else:  # If this was the second attempt, print an error message and continue with the next movie
                print(f'Request for {df["Title"][i]} failed: {e}')
                break
```
