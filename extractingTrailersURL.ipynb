```python
import googleapiclient.discovery
import pandas as pd
```

```python
def search(text):
    params = {
        "part": "snippet",
        "q": text,
        "type": "video",
        "maxResults": 1,
    }

    # Create a YouTube Data API v3 service object
    youtube = googleapiclient.discovery.build("youtube", "v3", developerKey="AIzaSyDshH7Xf2smqY__gDqF3qHA_NWwulAf7B0")

    # Call the search.list method
    results = youtube.search().list(**params).execute()

    # Get the URL of the first video
    video_url = f"https://www.youtube.com/watch?v={results['items'][0]['id']['videoId']}"

    # Print the video URL
    return (video_url)
```
